```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("Final preprocessing dataset.xlsx", header=1)
df.head()
```

```python
df.info()
```

```python
df['Year'] = pd.to_datetime(df['Year'], format='%Y')
df['Year'] = df['Year'].dt.year

```

```python
object_columns = df.select_dtypes(include=['object'])

# Convert the selected object columns to float
for col in object_columns.columns:
    df[col] = pd.to_numeric(df[col], errors='coerce', downcast='float')
```

```python
df.info()
```

```python
df.head()
```

```python
df['Unemployment Rate (%)'] = df['Unemployment Rate (%)'].apply(lambda x: x*100)
```

```python
# dropping the nominal gdp column because the real gdp represents a more accurate measure of the gdp

df = df.drop('Nominal GDP prices (Ksh Million)', axis=1)

df.head()


```

```python
df = df.iloc[:, :-3]
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have already calculated the correlation matrix
correlation_matrix = df.corr()

# Extract the correlation of 'Unemployment Rate (%)' with all other columns
unemployment_corr = correlation_matrix['Unemployment Rate (%)']

# Create a heatmap to visualize the correlations
sns.heatmap(unemployment_corr.to_frame(), annot=True, cmap='coolwarm')
plt.show()
```

```python
df.head()
```

```python
columns_to_fill = ['Urban population growth (annual %)', 'Adjusted net national income per capita (annual % growth)', 'Adjusted savings: education expenditure (current US$)']

# Iterate through each column and replace null values with the mean of that column
for column in columns_to_fill:
    column_mean = df[column].mean()
    df[column].fillna(column_mean, inplace=True)
    
df.isna().sum()
```

```python
df['Education_exp_ksh']  = df['Adjusted savings: education expenditure (current US$)'] * df['DOLLAR RATE ANNUALLY']

df.drop('Adjusted savings: education expenditure (current US$)', axis= 1, inplace=True)
df.head()
```

## `creating new features :`

## ` labor force to pop total`



```python
df['laborForce_toPOp_ration'] =round((df['Labour Force'] / df['Population, total'] *100),2)
df.head()
```

## `Inflation to gdp growth ratio`

```python
df['gdp_toInflation_rate'] =  round((df['Annual GDP growth (%)']/ df['Inflation Rate'] *100),2)

df.head()
```

## `Real gdp to labour force`

```python
df['realGDP_to_LaborForce(Ksh)'] =  df['Real GDP prices (Ksh Million)']/ df['Labour Force']

df.head()
```

```python
# Calculate the correlation matrix for your DataFrame (assuming 'df' is your DataFrame)
correlation_matrix = df.corr()

# Create a heatmap to visualize the correlations
plt.figure(figsize=(10, 8))  # Set the figure size for better readability
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")

# Add labels and title for the heatmap
plt.title("Correlation Heatmap")
plt.show()
```
